```python
import numpy as np
import sys, os
import glob
```

```python
filelist = glob.glob("./*.dat")
for file in filelist :
    base, _ = os.path.splitext(file)
    output = f"{base}.pb"
    with open(file) as f:
        lines = f.readlines()
        for k in range(len(lines)):
            l = lines[k]
            if l[0] == "#" : 
                continue
            else:
                idx = l.find(".")
                newline = l[:idx]+"0"+l[idx:]
                lines[k] = newline
        newf = open(output, "w")
        newf.writelines(lines)
```
